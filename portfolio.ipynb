```python
from transformers import pipeline
model_checkpoint = "timpal0l/mdeberta-v3-base-squad2/checkpoint-3603"
question_answerer = pipeline("question-answering", model=model_checkpoint)
def question_answer(context, question):
    predicted_answer = question_answerer(question=question, context=context)
    return predicted_answer['answer']
```

```python
#в данном блоке кода вы можете использовать свой текст и вопрос к нему чтобы получить предсказание
context = """Студия Epic Games анонсировала бета-тест шутера Gears of War 3. (2) Как сообщает GameSpot, тестированию подвергнется многопользовательский режим игры. (3) Сроки проведения бета-теста Gears of War 3 и требования для участия в нем объявлены не были. (4) Известно лишь, что он должен состояться до поступления игры в продажу. (5) Футуристический боевик Gears of War 3 был анонсирован 12 апреля 2010 года дизайнером студии Epic Games Клиффом Блезински (Cliff Bleszinski). (6) Игра разрабатывается для консоли Xbox 360. (7) Первоначально новая Gears of War должна была выйти в апреле 2011 года, однако в начале октября было объявлено, что игра отложена до осени 2011 года. (8) Первая часть Gears of War вышла 12 ноября 2006 года на на консоли Xbox 360. (9) PC-версия игры была выпущена 9 ноября 2007 года. (10) 7 ноября 2008 года на Xbox 360 вышла Gears of War 2. (11) Действие сериала происходит на планете Сера, где человечество отражает атаку подземной расы локустов. (12) Главные герои Gears of War - Маркус Феникс и Доминик Сантьяго, солдаты-ветераны, служащие в отряде \"Дельта\"."""
question = "Когда вышла первая часть Gears of War?"
question_answerer(question=question, context=context)
```

```python
#в этой ячейке мы распаковывем тестовую базу и выводим текст с вопросом предсказанием и возможными ответами
import pandas as pd    
jsonObj = pd.read_json(path_or_buf='test.jsonl', lines=True)
frame = pd.DataFrame(jsonObj['passage'])
for test_text in frame['passage']:
    test_text= pd.Series(test_text)
    question_answers = pd.Series(test_text['questions'])
    print('\n'+'\n'+'###########################################INPUT TEXT#############################################################')
    print(test_text['text'])
    for questions in question_answers:
        answers = [answers['text'] for answers in questions['answers']]
        print('\n'+'\n'+'###########################################INPUT QUESTION#############################################################')
        print(questions['question'])
        print('###########################################PREDICTED ANSWER#############################################################')
        print(question_answer(test_text['text'], questions['question']))
        print('###########################################POSSIBLE ANSWERS#############################################################')
        print(answers)
        
```
